```julia
using Plots, DataFrames, Flux
```

```julia
data = [24,26,29,22,18,16,12,20,22,26,23,18,14,12];
```

```julia
x = collect(UnitRange(1:length(data)));
```

```julia
pyplot()
```

```julia
data₁ = []
for i = 1:length(data)
    if i == 1
        push!(data₁, NaN)
    else
        push!(data₁, (data[i]-data[i-1])/1)
    end
end
```

```julia
data₂ = []
for i = 1:length(data₁)
    if i <= 2
        push!(data₂, NaN)
    else
        push!(data₂, (data₁[i]-data₁[i-1])/1)
    end
end
```

```julia
plot₁ = plot(x, data, ylims = (0,30), xticks = 0:1:14, lab = "Original Data")
plot₂ = plot(x, data₁, ylims = (-15,15), xticks = 0:1:14, lc = :orange, lab = "1st Derivative")
plot₃ = plot(x, data₂, ylims = (-15,15), xticks = 0:1:14, lc = :green, lab = "2nd Derivative")
plot(plot₁,plot₂,plot₃, layout = (3,1))
```

```julia
df = DataFrame()
df.x = x
df.data = data
df.data₁ = data₁
df.data₂ = data₂
df
```
